# Visualisation using matplotlib and seaborn

### matplotlib

<b>matplotlib is a plotting library for the Python programming language and its numerical mathematics extension NumPy. It provides an object-oriented API for embedding plots into applications using general-purpose GUI toolkits like wxPython, Qt, or GTK+. There is also a procedural "pylab" interface based on a state machine (like OpenGL), designed to closely resemble that of MATLAB. SciPy makes use of matplotlib.</b>

### seaborn

<b>Seaborn is a Python visualization library based on matplotlib. It provides a high-level interface for drawing attractive statistical graphics.</b>

### Adding some transparency will help

We can add transparency using the parameter <i>alpha</i> which takes values from zero (completely transparent - <i>i.e.</i>invisible!) to one (fully opaque - the default)

```python
plt.hist(dataset1,normed = True,color = 'blue',alpha = 0.2,bins = 10) 
plt.hist(dataset2,normed = True,color = 'red',alpha = 0.2,bins = 10)

# We normalise the data by setting normed = True 
# This is so that both can be plotted on the same scale
```

Setting the bandwidth manually

```python
for bw in np.arange(0.5,2,0.25):
    sns.kdeplot(dataset,bw = bw,label = bw,lw=1.8)
```

```python
webbrowser.open(url)
```

```python
kernel_options = ["biw", "cos", "epa", "gau", "tri", "triw"]
```

```python
for kern in kernel_options:
    sns.kdeplot(dataset,kernel = kern,label = kern,shade = True,alpha = 0.1,color = 'black')
```

```python
x3 = dframe_wine.alcohol
for kern in kernel_options:
    sns.kdeplot(x3,kernel = kern,label = kern,shade = True,alpha = 0.1,color = 'black')
```

## Multidimensional KDE plots

```python
mean = [0,0]
cov = [[1,0],[0,100]]
dataset2 = np.random.multivariate_normal(mean,cov,1000)
dframe = pd.DataFrame(dataset2,columns=['X','Y'])
```

```python
dframe.shape
```

```python
dframe.head()
```

```python
sns.kdeplot(dframe) # create multidimensional KDE plot of the data frame
```

```python
sns.kdeplot(dframe['X'],dframe['Y'])
```

Add shading

```python
sns.kdeplot(dframe.X,dframe.Y,shade = True)
```

```python
x1 = dframe_wine.citric_acid
x2 = dframe_wine.pH
```

```python
sns.jointplot(x1,x2)
```

```python
sns.kdeplot(x1,x2,shade = True)
```

```python
sns.kdeplot(dframe.X, dframe.Y,bw = 1,shade = True) # manually specify bandwidth
```

```python
sns.kdeplot(dframe.X,dframe.Y,bw = 'silverman') # automatically use Silverman's rule of thumb to estimate bw
```

```python
sns.jointplot('X','Y',dframe,kind = 'kde') # KDE as a joint plot
```

## Combining plot styles

```python
dataset = randn(1000)
```

```python
sns.distplot(dataset,bins = 25) # histogram with superimposed KDE plot; default bins for histogram = 10
```

```python
sns.distplot(dataset,bins = 25,rug = False,hist = False) # other options
```

Can set attributes for each of the plots separately by passing dicionaries to `distplot`

```python
sns.distplot(dataset,bins = 25,
            kde_kws={'color':'indigo','label':'KDE plot','alpha':0.9},
            hist_kws={'color':'orange','label':'Histogram'})
```

## Bar plots

A bar plot is roughly the discrete equivalent of a histogram, and is one of the most commonly used plot types

Let's create a small (fake) data set

```python
count = [23, 28, 34, 27, 24]
cities = ('New York', 'London', 'Tokyo', 'New Delhi', 'Sydney')
y_pos = np.arange(len(cities))
```

```python
# Create bars
plt.bar(y_pos, count)
```

To display the names on the x-axis, we need to add `xticks`

```python
plt.bar(y_pos, count)
plt.xticks(y_pos, cities)
```

```python
plt.bar(y_pos, count, color = 'green')
plt.xticks(y_pos, cities)
```

```python
plt.bar(y_pos, count, color = ['green','blue','red','orange','yellow'])
plt.xticks(y_pos, cities)
```

## Creating several subplots in the same plot

We can divide the plotting area into subplots using the `subplot()` function (which we used earlier to set the size of the plot).

```python
df = sns.load_dataset('iris')
```

```python
f, axes = plt.subplots(2,2,figsize=(10,8),sharex=False)
sns.distplot( df["sepal_length"] , color="blue", ax = axes[0, 0])
sns.kdeplot(df["sepal_width"] , color="green", ax = axes[0, 1])
sns.kdeplot(df["petal_length"] , color="brown", ax = axes[1, 0])
sns.distplot(df["petal_width"] , color="orange", ax = axes[1, 1])
```
